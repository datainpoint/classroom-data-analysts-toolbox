## 什麼是資料

- 資料是一個定義廣泛的名詞，又稱數據，是通過觀測得到的特徵或資訊。
- 資料存在不同的類別與形式，像是文字、數值、布林或者未定義值；當資料有關於多個人事時地物，就會以資料結構來組織。
- 更精準地說，如果我們希望透過電腦來處理資料，就會面對將觀測所得的特徵或資訊，轉換為不同類別的資料、並儲存於不同形式的資料結構。

## 什麼是分析

- 將某個有興趣的問題拆分為數個小問題。
- 研究小問題與有興趣的問題之間的關係，像是相關或者因果等。

## 什麼是「資料分析」

「資料分析」是一門透過邏輯、事實以及量化作為基準來拆解問題、並尋求解答的方法論。

## 為什麼透過「資料分析」處理問題

資料分析是連接問題與行動計畫之間的橋樑，基於邏輯、事實以及量化的行動計畫，我們才能夠對於問題的解決有樂觀預期。

1. 問題。
2. 資料分析。
3. 行動計畫。
4. 評估。
5. ...重複循環。

## 什麼是整合開發環境

> 寫作與執行程式需要三種類型的電腦軟體，包含純文字編輯器、終端機與直譯器（或編譯器），其中純文字編輯器能用來編寫程式、終端機能用來執行程式、直譯器（或編譯器）能將程式語言翻譯為電腦語言，整合開發環境就是將前述三者整合在一起的軟體。根據 2022 年的 Kaggle 調查，資料分析師的整合開發環境是 Jupyter Notebook、Visual Studio Code、PyCharm、RStudio 與 JupyterLab

來源：<https://en.wikipedia.org/wiki/Integrated_development_environment>

## 什麼是純文字編輯器

> 純文字編輯器主要用於編寫和檢視純文字檔案，因為程式與資料通常以純文字形式儲存，也就是不帶有格式（標題、段落或者項目符號等）的內容，所以程式設計師需要使用純文字編輯器來編輯程式、檢視資料。根據 2022 年的 Kaggle 調查，資料分析師的純文字編輯器是 Visual Studio Code

來源：<https://en.wikipedia.org/wiki/Text_editor>

## 什麼是模組、套件

> 程式語言組織程式碼的一種結構機制，透過模組、套件的開發管理，能夠實踐軟體工程中的 DRY(Don't Repeat Yourself)準則，現代軟體開發往往利用模組作為組裝單位，能分開地被編寫、允許同時協作。

來源：<https://en.wikipedia.org/wiki/Modular_programming>

## 什麼是商業智能軟體

> 視覺化分析軟體藉由 No-code、Low-code 的使用者圖形化介面，標榜以「拖、拉、點、選」讓人員能夠充分發揮資料的價值、改變組織使用資料解決問題的方式。根據 2022 年的 Kaggle 調查，資料分析師主要使用的商業智能軟體是 Tableau 與 Microsoft Power BI

來源：<https://en.wikipedia.org/wiki/Business_intelligence>

## 第零站：跟電腦變成好朋友

寫程式是理解資料分析、統計知識與機器學習理論不可或缺的工具與手段，因此學習的起點必須從程式設計起步，而在開始寫程式之前我們應該跟電腦變成好朋友！首先學習操作終端機以及純文字編輯器，行有餘力的話，再多認識 Git 和 GitHub。參考 [Data Science Foundations Tools and Techniques](https://www.amazon.com/gp/product/0135133106)

## 第一站：使用 SQL 查詢資料

從 2022 年的 Kaggle 調查我們得知資料分析師使用最多的語言依序是 Python -> SQL -> R，我會推薦從 SQL 開始，原因是 SQL 作為由資料庫中查詢資料的語言，與人類語言的相似度高、語法單純並且是即戰力的工具。參考我的好學校課程：[SQL 的五十道練習](https://hahow.in/cr/sqlfifty)。

懂得如何寫作 SQL 之後，可以跟試算表、商業智能軟體搭配開始進行探索性分析，從 2022 年的 Kaggle 調查我們得知 Tableau 與 Microsoft Power BI 是現在最普遍的商業智能軟體，而這兩個都有免費的版本可供個人電腦使用，如果是使用 Windows 作業系統的學員可以都試用看看，使用 macOS 的學員就試用 Tableau。

想要勝任工程導向的資料科學團隊職缺，必須學會一個在專案的各個應用都能介接、並且能夠自己整併清理資料的程式語言，從 2022 年的 Kaggle 調查我們得知資料分析師的首選是 Python 程式語言。參考我的好學校課程：[Python 的 50+ 練習](https://hahow.in/cr/pythonfiftyplus)。
